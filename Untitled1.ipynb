まずはアプリで採取したローカルデータ（快眠スコア）と、オープンデータ（睡眠指数）　を比較し、対応関係を調べる

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSVの読み込み
file_path = '/content/rocal_score.csv'  # アプリで計測した快眠スコアとサイトから引用した睡眠指数（1ヶ月分）
data = pd.read_csv(file_path)

# 視覚化
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='睡眠指数', y='快眠スコア', data=data)
plt.title('Relationship Between Sleep Index and Good Sleep Score')
plt.xlabel('Sleep Index')
plt.ylabel('Good Sleep Score')
plt.show()

correlation = data.corr()
print("Correlation Matrix:\n", correlation)
```

```python



file_path = '/content/rocal_score.csv'
data = pd.read_csv(file_path)

#対応関係の計算
average_scores = data.groupby('睡眠指数')['快眠スコア'].mean()

#快眠スコアと睡眠指数の対応関係を表示
print("Average Good Sleep Score for Each Sleep Index:")
print(average_scores)
```

#結果
睡眠指数が1の場合、快眠スコアは平均５６点。
睡眠指数が２の場合、快眠スコアは平均６１点。
睡眠指数が3の場合、快眠スコアは平均71点。
睡眠指数が4の場合、快眠スコアは平均79点。
睡眠指数が5の場合、快眠スコアは平均82点。





#次に快適な室内環境における快眠スコアを取得し、上記の結果をもとに睡眠指数に換算する

```python


# 以前のデータセットから平均スコアを再計算するためのコード
def calculate_average_scores():
    # 以前のデータセットを読み込む
    file_path_previous = '/content/rocal_score.csv'  # 以前のCSVファイルへのパス
    data_previous = pd.read_csv(file_path_previous)
    return data_previous.groupby('睡眠指数')['快眠スコア'].mean()

# 快眠スコアを睡眠指数に変換するための関数
def find_closest_sleep_index(score, average_scores):
    differences = abs(average_scores - score)
    return differences.idxmin()

# 新しいデータセットを読み込む
file_path_rocal = '/content/rocal2.csv'  # 新しいCSVファイルへのパス
data_rocal = pd.read_csv(file_path_rocal)

# 平均スコアを計算
average_scores = calculate_average_scores()

# 快眠スコアを睡眠指数に変換
data_rocal['Calculated 睡眠指数'] = data_rocal['快眠スコア'].apply(lambda score: find_closest_sleep_index(score, average_scores))

# 結果を表示
print(data_rocal)
```

計算の結果、快眠スコアを睡眠指数に換算することに成功

#最後に、オープンデータからなる睡眠指数と先ほど換算したローカルデータを視覚化して比較

```python

# ファイルを読み込む
file_path_conmare = '/content/compare.csv'  # conmare.csvファイルへのパス
conmare_data = pd.read_csv(file_path_conmare)

# 視覚化の設定
plt.style.use('seaborn-darkgrid')

# 比較グラフの作成
plt.figure(figsize=(12, 6))
plt.plot(conmare_data['日付'], conmare_data['ローカル'], label='ローカル (Local)', marker='o')
plt.plot(conmare_data['日付'], conmare_data['オープン'], label='オープン (Open)', marker='x')
plt.title('Comparison of Local and Open Sleep Index Over Time')
plt.xlabel('Date')
plt.ylabel('Sleep Index')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

グラフからわかるように、やはりtenki.jpで公開されているオープンデータはあくまで外気温をもとに設定しているものであるから、室内で生活する我々にとって良い判断の指標にはならないことがわかった。睡眠指数が5段階でしか評価できないため数値上はオープンとローカルで一致する場合があったが、点数設定をさらに細分化することができたら、その差は如実なものになったかもしれない。
